```python
# page2 and onward
HOST = 'https://udn.com'
pg2_url_tmpl = "https://udn.com/news/get_breaks_article/{}/1/99"

res_links = []

for pg_num in range(2,11):
    pg = r.get(pg2_url_tmpl.format(pg_num))
    pg.encoding = 'utf-8' # 調整Response obect的encoding，不然中文是亂碼
    soup = BeautifulSoup(pg.text, 'html5lib')
    res_links.extend([HOST + link['href'].split("?")[0] for link in soup.select("h2 > a")])
```

```python
def str2dict(mystr):
    res_dict = {}
    for line in mystr.split('\n'):
        res_dict[line.split(":")[0]] = ":".join(line.split(":")[1:])
    return res_dict
```

```python
import urllib.parse

pg = 2

while True:
    URL = "https://web.pcc.gov.tw/tps/pss/tender.do"

    query_strings = {
        "searchMode": "common",
        "searchType": "advance",
        "searchTarget": "ATM",
        "method": "search",
        "isSpdt": "",
        "pageIndex": pg

    }

    res = s.get(URL, params=query_strings)
    
    # 如果頁面上出現無符合條件，跳出while loop
    if "無符合條件資料" in res.text:
        break
    
    with open("./pccweb/{}.html".format(pg), 'w') as f:
        f.write(res.text)
    pg += 1
    
```
